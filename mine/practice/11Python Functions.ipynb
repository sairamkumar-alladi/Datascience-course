```python
# Function Definition
def add(a,b):# Function Inputs
    #Function Logic
    c = a+b
    # Return Logic
    return c
```

```python
result = add(1,2) #Function Calling
```

```python
def display_output(result):
    print(f"Result is: {result}")
```

```python
display_output(add(1,2)) # We are calling a function inside a function call
```

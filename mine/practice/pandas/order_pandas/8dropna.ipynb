The df.dropna(thresh=1) function in pandas is used to remove rows from a DataFrame that have at least one non-null (non-NaN) value.

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with NaN values
data = {
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, np.nan],
    'C': [1, np.nan, np.nan, 4]
}

df = pd.DataFrame(data)

# print("Original DataFrame:")
# print(df)

# # Drop rows with at least one non-null value
# df_cleaned = df.dropna(thresh=1)

# print("\nDataFrame after dropping rows with at least one non-null value:")
# print(df_cleaned)
df
```

```python
# Drop rows with at least one non-null value
df_cleaned = df.dropna(thresh=1)

print("\nDataFrame after dropping rows with at least one non-null value:")
print(df_cleaned)
```
